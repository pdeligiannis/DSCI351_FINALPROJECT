```python
#I need to transpose the data for the cufflinks library to work that's how it works as per documentation 
def create_returns(table,list_of_tickers,transpose_id): 
    aggregate_returns_df = table[['returns','id']]
    returns_df = pd.DataFrame(index=securities_table['Date'].unique().tolist(), columns=list_of_tickers.values())
    i = 0 
    for column in returns_df.columns: 
        returns_df[column] = aggregate_returns_df.loc[aggregate_returns_df['id']==i][transpose_id].tolist()
        i += 1 
    returns_df = returns_df.dropna() 
    return returns_df
```

```python
#show tables function 
def showtable(cur,tables):
    cur.execute("SELECT * FROM "+tables)
    rows = cur.fetchall()
    names = list(map(lambda x: x[0], cur.description))
    show_table = pd.DataFrame(columns=names)
    show_table.rows = rows 
    for i in range(len(rows)): 
        show_table.loc[i] = rows[i]
    return show_table
```

```python
#retrieve id -> tickers mapper 
cur.execute("SELECT distinct id,ticker from company")
rows = cur.fetchall() 
list_of_tickers = {}
for row in rows: 
    list_of_tickers[row[0]] = row[1]
```

```python
returns_df = create_returns(securities_table,list_of_tickers,"returns")
cf.set_config_file(sharing='public',theme='pearl',offline=False)
cf.go_offline()
returns_df.iplot(filename='Returns of Equities',title='Stock Returns')
```

```python
cumulative_returns_df = returns_df.cumsum()
cumulative_returns_df.iplot(filename='Cumulative Returns of Equities',title='Stock Cumulative Returns')
```

```python
#create search function based on different possible queries 
def search_function(cur,query):  
    #create query 
    query = query.strip() 
    cur.execute(query)
    names = list(map(lambda x: x[0], cur.description))
    search_result = pd.DataFrame(columns=names)
    rows = cur.fetchall() 
    for i in range(len(rows)):
        search_result.loc[i] = rows[i]
    return search_result
```

```python
#SEARCH FOR COMPANY TABLE 

column_company = 'ticker,name,city'
table = 'company'
query = 'select '+column_company+' from '+table+' '

keyword = 'Mountain View'
case='1'

if case=='0':
    query = query + "where city= '"+keyword+"' "
if case=='1':
    query = query + "where city like '%" + keyword + "%' "

# * WHERE CONDITION 
# * WHERE LIKE CONDITION "where column like '%" + keyword + "%' "
# * GROUP BY CONDITION 
# * JOIN CONDITION?? 

search_result = search_function(cur,query)
search_result
```

```python
#SEARCH FOR COMPANY TABLE 

column_company = 'adj_close,id'
table = 'security_price'
query = 'select '+column_company+' from '+table+' '

case='2'

#*** MAKE THIS MORE OF A DYNAMIC SEARCH RESULT ... don't hardcore the query 
#*** create an architectural flow to present that tomorrow! 

if case=='0':
    query = query + "where id = 0"
if case=='1': #double where condition
    query = query + "where adj_close>=120 and id=0"
if case=='2':
    query = "select id, adj_close, avg(adj_close) from security_price where id=0" 

# * WHERE CONDITION 
# * WHERE CONDITION AND HAVING 

search_result = search_function(cur,query)
search_result
```
